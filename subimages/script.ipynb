```python
from PIL import Image
import os

# Directorio que contiene las imágenes
directorio_imagenes = './'

# Resolución deseada
resolucion_deseada = (106, 103)

def eliminar_imagen(imagen_path):
    os.remove(imagen_path)
    print(f"Imagen eliminada: {imagen_path}")

def verificar_resolucion(imagen_path, resolucion_deseada):
    try:
        imagen = Image.open(imagen_path)
        return imagen.size == resolucion_deseada
    except Exception as e:
        print(f"Error al abrir la imagen {imagen_path}: {e}")
        return False

def eliminar_imagenes_con_resolucion_incorrecta(directorio, resolucion_deseada):
    for filename in os.listdir(directorio):
        filepath = os.path.join(directorio, filename)
        if os.path.isfile(filepath):
            if not verificar_resolucion(filepath, resolucion_deseada):
                eliminar_imagen(filepath)
```

```python
eliminar_imagenes_con_resolucion_incorrecta(directorio_imagenes, resolucion_deseada)
```

```python
import os

directorio = "./noforest/"  # Reemplaza esto con la ruta de tu directorio de imágenes

# Obtén la lista de archivos en el directorio
archivos = os.listdir(directorio)

# Filtra solo los archivos con extensiones de imagen (puedes ajustar esto según tus necesidades)
archivos_de_imagen = [archivo for archivo in archivos if archivo.lower().endswith((".jpg", ".jpeg", ".png", ".gif"))]

# Ordena los archivos para garantizar un orden consistente
archivos_de_imagen.sort()

# Itera sobre los archivos y renómbralos
for i, archivo in enumerate(archivos_de_imagen, 1):
    # Construye el nuevo nombre de archivo
    nuevo_nombre = f"noforest_{i}{os.path.splitext(archivo)[1]}"  # Conserva la extensión original
    
    # Construye la ruta completa de origen y destino
    ruta_origen = os.path.join(directorio, archivo)
    ruta_destino = os.path.join(directorio, nuevo_nombre)
    
    # Renombra el archivo
    os.rename(ruta_origen, ruta_destino)

print("Proceso completado.")
```
